##### Membuat Model Regresi Linear

1. Import Library

```python
# Import library pandas
import pandas as pd

# Import library numpy
import numpy as np

# Import library matplotlib dan seaborn untuk visualisasi
import matplotlib.pyplot as plt
import seaborn as sns

# Import modul LinearRegression digunakan untuk memanggil algoritma Linear Regression
from sklearn.linear_model import LinearRegression

# Import modul train test split digunakan untuk membagi data kita menjadi training dan testing set
from sklearn.model_selection import train_test_split

# Inport Math agar program dapat menggunakan semua fungsi yang ada pada modul math (ex:sqrt)
import math

# Me-non aktifkan peringatan pada Python
import warnings
warnings.filterwarnings('ignore')
```

2. Cek Dataset

```python
data =  'Dataset/CarPrice_Assignment.csv'
dataset = pd.read_csv(data)
dataset.head(10)
```

```python
dataset.info()
```

```python
dataset.isna().sum()
```

3. Cek Korelasi

```python
dataset.corr()
```

```python
def pp(x,y,z):
    sns.pairplot(dataset, x_vars=[x,y,z], y_vars='price', size=4, aspect=1, kind='scatter')
    plt.show()

pp('enginesize', 'boreratio', 'stroke')
pp('compressionratio', 'horsepower', 'peakrpm')
pp('wheelbase', 'citympg', 'highwaympg')
```

```python
plt.scatter(dataset['enginesize'], dataset['price'])
plt.xlabel('enginesize')
plt.ylabel('price')
plt.title('Scatter Plot enginesize vs price')
plt.show()
```

4. Membangun Model

```python
# Mengubah bentuk matriks dengan reshape
x = dataset['enginesize'].values.reshape(-1,1)
y = dataset['price' ].values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state = 100)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
# Bagian ini untuk mengecek nilai koefisien dan intersep
print(regressor.coef_)
print(regressor.intercept_)
```

```python
# Cek score model
regressor.score(X_test,y_test)
```

```python
# Bagian ini bisa diskip
print(X_test)
```

```python
# Prediksi data test
y_prediksi = regressor.predict(X_test)

plt.scatter(X_test,y_test)
plt.plot(X_test, y_prediksi, c='r')
plt.xlabel('enginesize')
plt.ylabel('Price')
plt.title('Plot enginesize vs Price')
```

```python
# Prediksi harga mobil dengan nilai enginesize tertentu

print('Nilai prediksi harga dengan enginesize 100 : ',regressor.predict([[90]]))
print('Nilai prediksi harga dengan enginesize 150 : ',regressor.predict([[150]]))
print('Nilai prediksi harga dengan enginesize 200 : ',regressor.predict([[200]]))
```
